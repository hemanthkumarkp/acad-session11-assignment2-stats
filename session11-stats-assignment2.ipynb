```python
import numpy as np
import statistics as stats
assignmentMarks = [6, 7, 5, 7, 7, 8, 7, 6, 9, 7, 4, 10, 6, 8, 8, 9, 5, 6, 4 ,8]

print("Input Array in ascending order: ", np.sort(assignmentMarks))

print("Mean of assignmentMarks ", assignmentMarks," is: ", np.mean(assignmentMarks))
print("Median of assignmentMarks ", assignmentMarks," is: ", np.median(assignmentMarks))
print("Mode of assignmentMarks ", assignmentMarks," is: ", stats.mode(assignmentMarks))
```
